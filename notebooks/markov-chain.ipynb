```python
# !pip install basix
from basix.viz import define_colormap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

labels = ["Início", "Busca", "Facebook", "Google Ads", "Desistência", "Conversão"]
palette = ["#000000", "#548235", "#2F5597", "#C00000", "#7030A0", "#E2770C"]

M = np.array(
    [[0,0,0,0,0,0], 
     [0.5,0.16667,0.125,0.25,0,0],
     [0.25,0.25,0,0.25,0,0],
     [0.25,0.25,0.125,0,0,0],
     [0, 0.16667, 0.5, 0.25, 1, 0],
     [0, 0.16667, 0.25, 0.25, 0, 1]]
)

S0 = np.array([[1],[0],[0],[0],[0],[0]])

states = [S0.flatten()]
for i in range(6):
    state = np.matmul(M,states[-1])
    states.append(state.flatten())

states = np.array(states)

for i in range(1,6):
    plt.plot(states[:,i], c=palette[i], label = labels[i])
    
plt.legend()
plt.xlabel("Períodos")
plt.ylabel("Probabilidade")
plt.show()
```
