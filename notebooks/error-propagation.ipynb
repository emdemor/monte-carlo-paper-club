```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# lifetime values
ltv_m = 200
ltv_s = 70

# cac
cac_m = 100
cac_s = 60

# Sampling ROI
n = 10000
sample_ltv = ltv_m + ltv_s * np.random.randn(n) # LTV ~ N(200,70²)
sample_cac = cac_m + cac_s * np.random.randn(n) # CAC ~ N(100,60²)
roi = sample_ltv-sample_cac

# Outputs
plt.hist(roi, bins=50)
plt.xlabel('ROI')

print('E|ROI| = R$ ({:.2f} +- {:.2f})'.format(np.mean(roi), np.std(roi)))
print("PCT|ROI>0| = {:.0f}%".format(100*np.mean(roi > 0)))
plt.show()
```
