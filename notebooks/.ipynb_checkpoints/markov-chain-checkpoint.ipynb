```python
# !pip install basix
from basix.viz import define_colormap
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

colors = ['#ffc801ff','#ffc800','#1bcbdcff','#0073cd',]
cmap = define_colormap(colors)
```

```python
M = np.array(
    [[0,0,0,0,0,0], 
     [0.5,0.16667,0.125,0.25,0,0],
     [0.25,0.25,0,0.25,0,0],
     [0.25,0.25,0.125,0,0,0],
     [0, 0.16667, 0.5, 0.25, 1, 0],
     [0, 0.16667, 0.25, 0.25, 0, 1]]
)

S0 = np.array([[1],[0],[0],[0],[0],[0]])


states = [S0.flatten()]

for i in range(6):
    state = np.matmul(M,states[-1])
    states.append(state.flatten())

states = np.array(states)
```

```python
import matplotlib.pyplot as plt

plt.plot(states[:,1], c="#548235", label = "Busca")
plt.plot(states[:,2], c="#2F5597", label = "Facebook")
plt.plot(states[:,3], c="#C00000", label = "Google Ads")
plt.plot(states[:,4], c="#7030A0", label = "Desistência")
plt.plot(states[:,5], c="#E2770C", label = "Conversão")
plt.legend()
plt.show()
```
