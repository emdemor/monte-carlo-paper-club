## Função considerada

```python
f = lambda x: ((np.sin(10*x**2))**2*np.sin(x))*x+0.1

X_true = np.linspace(0,1,1000)
y_true = f(X_true)

plt.plot(X_true, y_true)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0,0.8])
plt.savefig('f.svg', transparency=True)
plt.show()
```

## Pontos

```python
K = 10000

x_range = [0,1]
y_range = [0,0.8]

X_rand = np.random.uniform(x_range[0],x_range[1],K)
f_rand = f(X_rand)
y_rand = np.random.uniform(0, y_range[1],K)

Area = (y_range[1]-0) * (x_range[1] - x_range[0])

points_in = y_rand < f_rand
print(points_in.sum())
print('Area Estimate:', Area * points_in.mean())
```

## Points

```python
plt.scatter(X_rand, y_rand, marker='.', s=3, c='black',alpha=0.2)
plt.plot(X_true, y_true)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0,0.8])
plt.savefig('../figures/fp.png', transparent=True, dpi=300)
plt.show()
```

```python
plt.scatter(X_rand[points_in], y_rand[points_in], marker='.', s=3, c=colors[3],alpha=0.4)
plt.scatter(X_rand[~points_in], y_rand[~points_in], marker='.', s=3, c=colors[1],alpha=0.4)

plt.plot(X_true, y_true)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([0,0.8])
plt.savefig('fp_c.png', transparent=True, dpi=300)
plt.show()
```

## Error

```python
def monte_carlo_estimator(f, k, x_range = [0,1], y_range=[0,0.8]):
    X_rand = np.random.uniform(x_range[0],x_range[1],k)
    f_rand = f(X_rand)
    y_rand = np.random.uniform(0,y_range[1],k)
    area = (y_range[1]-0) * (x_range[1] - x_range[0])
    points_in = y_rand < f_rand
    return area * points_in.mean()
```

```python
from tqdm import tqdm
res = np.array(
    [
        (
            int(round(k)), 
            np.abs(monte_carlo_estimator(f, k=int(round(k)), x_range = [0,1])-0.242633)
        ) for k in tqdm(np.logspace(1,7,200))
    ]
)
```

```python
X_plot = np.logspace(1,7,7)

y_plot = 0.2/np.sqrt(X_plot)

plt.loglog(res[:,0], res[:,1], label="erro na estimativa")
plt.loglog(X_plot, y_plot, label = "1 / √N")
plt.legend()
plt.xlabel("Número de Pontos")
plt.show()
```
